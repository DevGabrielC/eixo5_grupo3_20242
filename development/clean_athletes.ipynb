```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
results = pd.read_csv('../datalake/bronze/athlete_results.csv')

df = results.copy()
```

```python
df.head()
```

```python
# Take off the "(Olympic)" and other edition types inside Event column
df['edition_type'] = df['Event'].str.extract(r'\((.*?)\)')
df['Event'] = df['Event'].str.replace(r'\s*\(.*?\)', '', regex=True)

# Get place column
df['place'] = df['Pos'].str.extract(r'(\d+)')
```

```python
df[['year', 'type']] = df['Games'].str.extract(r'(\d{4}) (Summer|Winter)', expand=True)
df['year'] = pd.to_numeric(df['year'])
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```
